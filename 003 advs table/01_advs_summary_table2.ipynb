# 使用advs来生成Summary Table

```python
import pandas as pd
import numpy as np
from plotnine import *
%matplotlib inline


advs = pd.read_sas("./data/processed/advs.xpt")
```

```python
from pathlib import Path
from datetime import datetime
```

```python
advs1 = advs.loc[(advs["ANL01FL"] == b"Y") & (advs["PARAM"] == b'Diastolic Blood Pressure (mmHg)')]
gg = advs1.groupby(["TRTPN","TRTP","AVISITN","AVISIT"])
```

```python
advs1.head()
```

```python
gg["PARAM"].value_counts()
```

```python
sum = gg["AVAL"].agg({"count","mean","std","sem","median","min","max"})
```

```python
list(sum)
```

```python
sum["mean"] = sum["mean"].round(1).apply(str)
sum["std"] = sum["std"].round(2).apply(str)
sum["sem"] = sum["sem"].round(2).apply(str)
sum["median"] = sum["median"].round(1).apply(str)
sum["min"] = sum["min"].round(0).apply(str)
sum["max"] = sum["max"].round(0).apply(str)
sum.head()
```

```python
sum.dtypes
```

```python
sum.reset_index(inplace=True)
sum.sort_values(by=["AVISITN","AVISIT","TRTPN","TRTP"])
```

```python
sum_t=pd.melt(sum,id_vars=["AVISITN","AVISIT","TRTPN","TRTP"],var_name='stats', value_name='valn')
```

```python
sum_t.head()
```

```python
sum_t.dtypes
```

```python
# refer to http://pbpython.com/pandas_dtypes.html
sum_t["AVISITN"] = sum_t["AVISITN"].astype("int")
sum_t["TRTPN"] = sum_t["TRTPN"].astype("int")
```

```python
sum_t.dtypes
```

```python
sum_t.head()
```

```python
sum_t.loc[sum_t["stats"] == "count", "statn"] = 1
sum_t.loc[sum_t["stats"] == "mean", "statn"] = 2
sum_t.loc[sum_t["stats"] == "median", "statn"] = 3
sum_t.loc[sum_t["stats"] == "std", "statn"] = 4
sum_t.loc[sum_t["stats"] == "sem", "statn"] = 5
sum_t.loc[sum_t["stats"] == "min", "statn"] = 6
sum_t.loc[sum_t["stats"] == "max", "statn"] = 7
```

```python
sum_t.loc[sum_t["stats"] == "count", "statlabel"] = "n"
sum_t.loc[sum_t["stats"] == "mean", "statlabel"] = "Mean"
sum_t.loc[sum_t["stats"] == "median", "statlabel"] = "Median"
sum_t.loc[sum_t["stats"] == "std", "statlabel"] = "S.D."
sum_t.loc[sum_t["stats"] == "sem", "statlabel"] = "S.E."
sum_t.loc[sum_t["stats"] == "min", "statlabel"] = "Min"
sum_t.loc[sum_t["stats"] == "max", "statlabel"] = "Max"
```

```python
sum_t.head()
```

```python
sum_t["varn"] = 1
sum_t["var"] = "AVAL"
sum_t["varlabel"] = b'Diastolic Blood Pressure (mmHg)'
sum_t.head()
```

```python
#final=sum_t.pivot_table(index=["AVISITN","AVISIT","varn","var","varlabel","statn","stats","statlabel"], columns=["TRTPN","TRTP"], values='valn').reset_index()
#final=sum_t.pivot(index=["AVISITN","AVISIT","varn","var","varlabel","statn","stats","statlabel"], columns=["TRTP"]).reset_index()
#final.head()

sum_t = sum_t.drop(columns=["TRTPN"])
sum_t = sum_t.set_index(["AVISITN","AVISIT","varn","var","varlabel","statn","stats","statlabel","TRTP"])
```

```python
list(sum_t)
```

```python
final = sum_t.unstack().reset_index()
final.head()
```

```python
list(final)
```

```python
# export to html
final.to_html("./reports/01_advs_diabp_aval_sum.html")
```
